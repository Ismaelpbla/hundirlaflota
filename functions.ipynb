```python
import numpy as np
import random as rd

dimensiones = (10, 10)
barcos = {'A': np.array([0,0]) , 'B':np.array([0,0]) , 'C': np.array([0,0]), 'D': np.array([0,0]), 'AA': np.array([(0,0),(0,0)])  , 'BB': np.array([(0,0),(0,0)]),  'CC': np.array([(0,0),(0,0)]), 'AAA': np.array([(0,0,0),(0,0,0)]), 'BBB': np.array([(0,0,0),(0,0,0)]), 'S': np.array([(0,0,0,0),(0,0,0,0)])}
tablero_jugador = np.full((10,10),'')
tablero_maquina = np.full((10,10),'')
tablero_check = np.full((11,11),'')
orientation_list = ['v','h']

def generacion_aleatoria (boats):

    for key,values in boats.items():
        if key == 'A' or key == 'B' or key == 'C' or key == 'D':
            values = np.array([(np.random.randint(0,10),np.random.randint(0,10))])
            print(values)
            boats[key] = values
        elif key == 'AA' or key =='BB' or key =='CC':
            orientation = rd.choice(orientation_list)
            print(orientation)
            if orientation == 'v':
                a = np.random.randint(0,9)
                b = np.random.randint(0,9)
                values = np.array([(a,a),(b,b+1)])
                print(values)
                boats[key] = values
            else:
                c = np.random.randint(0,9)
                d = np.random.randint(0,9)
                values = np.array([(c,c+1),(d,d)])
                print(values)
                boats[key] = values
        elif key == 'AAA' or key == 'BBB':
            orientation = rd.choice(orientation_list)
            print(orientation)
            if orientation == 'v':
                e = np.random.randint(0,8)
                f = np.random.randint(0,8)
                values = np.array([(e,e,e),(f,f+1,f+2)])
                print(values)
                boats[key] = values
            else:
                g = np.random.randint(0,8)
                h = np.random.randint(0,8)
                values = np.array([(g,g+1,g+2),(h,h,h)])
                print(values)
                boats[key] = values
        elif key == 'S':
            orientation = rd.choice(orientation_list)
            print(orientation)
            if orientation == 'v':
                k = np.random.randint(0,7)
                l = np.random.randint(0,7)
                values = np.array([(k,k,k,k),(l,l+1,l+2,l+3)])
                print(values)
                boats[key] = values
            else:
                m = np.random.randint(0,7)
                n = np.random.randint(0,7)
                values = np.array([(m,m+1,m+2,m+3),(n,n,n,n)])
                print(values)
                boats[key] = values
    return boats


def dist(tupa, tupb):
    return np.sqrt((tupa[0]-tupb[0])**2 + (tupa[1]-tupb[1])**2)

failed_board = True

while failed_board == True:
    failed_board = False
    print('hasta aqui')
    d = generacion_aleatoria(barcos)
    lista_barcos = list(d.values())    

    for i in range(len(lista_barcos)):
        control0 = False
        barco1 = lista_barcos[i]
        for barco2 in lista_barcos[i+1:]:
            control1 = False
            for tupla1 in barco1:
                control2 = False
                for tupla2 in barco2:
                    control3 = False
                    dis = dist(tupla1,tupla2)
                    if dis < 2:
                        control3 = True
                        break
                if control3 == True:
                    control2 = True
                    break
            if control2 == True:
                control1 = True
                break
        if control1 == True:
            control0 = True
            break
    if control0 == True:
        failed_board = True
else:
    print('se ha conseguido')
   


```
